# Create Model

use tf and keras both

design the nural network . three layers. CNN - convoluntial Nural network. this is a 2x2 numpy arrau as input. it is a platten layer. 
activation function is relu.
output layer - here it is 10 as we need to get a digit out put

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(10)
])

print(model.summary())
```

need to traing 101,770 parametrs

# Compile Model

we aare using adam optimiser. more optimsers are availble in keras library 

losss function decide error / 

metrics- which metric should be used to calculate the model

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=[tf.keras.metrics.SparseCategoricalAccuracy()],
)
```

# Fit Model

number of epochs
need more epochs for better model

```python
# Fit model
hist = model.fit(X_train, y_train,
                batch_size=128,
                epochs=10,
                verbose=1,
                validation_data=(X_test, y_test))
```

# Model Fitting History

```python
fit_history = pd.DataFrame(hist.history)
fit_history.head()
```

```python
fit_history[['sparse_categorical_accuracy', 'val_sparse_categorical_accuracy' ]].plot()
```

# Evaluate Model

```python
score = model.evaluate(X_test, y_test, verbose=0)    
print(model.summary())
print('train samples:', X_train.shape[0])
print('test samples:', X_test.shape[0])
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

# Convolution Neural Network (CNN) Model

## Load Dataset and Normalize

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X_train, X_test = X_train / 255.0, X_test / 255.0

pyplot.imshow(X_test[0])
pyplot.colorbar()
print("Number = ", y_test[0])
```

## Transform image data into single channel (monochrome)

```python
X_train = X_train.reshape((X_train.shape[0], 28, 28, 1))
X_test = X_test.reshape((X_test.shape[0], 28, 28, 1))
```

```python
print("Test Dataset:", X_test.shape)
print("Train Dataset:", X_train.shape)
```

### One-hot encode y variable

```python
y_train = tf.keras.utils.to_categorical(y_train)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
y_train
```

```python
print("y_test:", y_test.shape)
print("y_train:", y_train.shape)
```

## CNN Model

```python
input_shape = (28, 28, 1)
num_classes = 10

model = tf.keras.Sequential(
    [
        tf.keras.Input(shape=input_shape),
        tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation="relu"),
        tf.keras.layers.MaxPooling2D(pool_size=(2, 2)),
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dropout(0.5),
        tf.keras.layers.Dense(10, activation="softmax"),
    ]
)

model.summary()
```

```python
batch_size = 128
epochs = 10

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

model.fit(X_train, y_train, 
          batch_size=batch_size, 
          epochs=epochs, 
          validation_split=0.1, 
          validation_data=(X_test, y_test))
```

## Evaluate Model

```python
score = model.evaluate(X_test, y_test, verbose=0)    
print(model.summary())
print('train samples:', X_train.shape[0])
print('test samples:', X_test.shape[0])
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

## Confusion Matrix

```python
test_predictions = model.predict(X_test)   
test_predictions
```

```python
# Decide the class based on the class with maxiumum predicted value.
test_predictions = np.argmax(test_predictions, axis=1)
test_predictions
```

```python
# Convert One hot encoded y column to categorical column
y_test = np.argmax(y_test, axis=1)
y_test
```

```python
test_result = pd.DataFrame({'y_act': y_test, 'y_pred':test_predictions})
test_result.sample(5)
```

```python
cfm = pd.crosstab(test_result['y_act'], test_result['y_pred'], margins=True)
cfm
```
